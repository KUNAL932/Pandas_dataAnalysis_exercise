### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Create 3 differents Series, each of length 100, as follows: 
1. The first a random number from 1 to 4 
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
series1 = pd.Series(np.random.randint(1,5,100))
series2 = pd.Series(np.random.randint(1,4,100))
series3 = pd.Series(np.random.randint(10000,30001,100))
print(series1 ,series2,series3)
```

### Step 3. Let's create a DataFrame by joinning the Series by column

```python
new_df = pd.concat([series1,series2,series3],axis = 1)
new_df
```

```python
new_df = new_df.rename(columns = {0:'bedrs',1:'bathrs',2:'price_sqr_meter'})
new_df.head()
```
